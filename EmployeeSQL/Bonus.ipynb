```python
from sqlalchemy import create_engine
import getpass
USR = 'postgres'
PWD = getpass.getpass()
```

```python
SQLALCHEMY_DATABASE_URI = 'postgresql://{}:{}@localhost:5432/Company'.format(USR, PWD)
engine = create_engine(SQLALCHEMY_DATABASE_URI)
connection = engine.connect()
```

```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
salaries = connection.execute("Select * from salaries").fetchall()
sal_df = pd.DataFrame(salaries)
```

```python
plt.hist(sal_df[2])
plt.title("Income by Frequency")
plt.xlabel("Income")
plt.ylabel("Frequency")
```

```python
salary_by_title = connection.execute("Select T.title, AVG(S.salary) FROM SALARIES S JOIN EMPLOYEES E ON E.emp_no = S.emp_no JOIN TITLES T ON T.emp_no = E.emp_no GROUP BY T.title")
```

```python
salary_by_title = salary_by_title.fetchall()
```

```python
sal_by_title = pd.DataFrame(salary_by_title)
```

```python
plt.bar(x = sal_by_title[0], height = sal_by_title[1])
plt.xticks(rotation="vertical")
plt.title("Income by Job Type")
plt.ylabel("Income ($)")
```
